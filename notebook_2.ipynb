```python
import pandas as pd
import numpy as np
from datetime import datetime

historical_data = pd.read_csv('/content/drive/MyDrive/ds_supriya mogarala/csv_files/historical_data.csv')

historical_data['date']=pd.to_datetime(
    historical_data['Timestamp IST'],
    format='%d-%m-%Y %H:%M'
).dt.date
historical_data['Coin']=historical_data['Coin'].str.replace('@', '')

fear_greed_index =pd.read_csv('/content/drive/MyDrive/ds_supriya mogarala/csv_files/fear_greed_index.csv')
fear_greed_index['date']=pd.to_datetime(fear_greed_index['date']).dt.date
```

```python
merged_data = pd.merge(historical_data,fear_greed_index,on='date',how='left')
```

```python
Trade_metrics = merged_data.groupby(['date','classification']).agg(
    Trade_volume=('Size USD','sum'),
    Profit=('Closed PnL','sum'),
    Trade_size=('Size USD','mean'),
    Fees=('Fee','sum'),
    Trade_count=('Trade ID','count')
).reset_index()
Trade_metrics
```
